Distribution of scores over all games

```python
#Ok so this whole mess makes a new column, 'score bins', games in the top third get a value of 3, in the second third 2, and bottom third 1

threshold=int(len(dfonehot)/3)

a=dfonehot.sort_values(by=['User Score'],axis=0, ascending=False).iloc[threshold-1:threshold]['User Score']
dfonehot['High Score']=dfonehot['User Score']>a.values[0]
dfonehot["High Score"] = dfonehot["High Score"].astype(int)

b=dfonehot.sort_values(by=['User Score'],axis=0, ascending=False).iloc[threshold*2-1:threshold*2]['User Score']
dfonehot['Average Score']=dfonehot['User Score']>b.values[0]
dfonehot["Average Score"] = dfonehot["Average Score"].astype(int)

c=dfonehot.sort_values(by=['User Score'],axis=0, ascending=False).iloc[threshold*3-1:]['User Score']
dfonehot['Low Score']=dfonehot['User Score']>c.values[0]
dfonehot["Low Score"] = dfonehot["Low Score"].astype(int)

dfonehot['Score Bins']=dfonehot['High Score']+dfonehot['Average Score']+dfonehot['Low Score']
dfonehot=dfonehot.drop(['High Score','Average Score','Low Score'],axis=1)
dfonehot['Score Bins']
#Getting only the non numeric info
dfonehot=dfonehot.select_dtypes(include=np.number)
#drops rows with NANs
dfonehot=dfonehot.dropna()
#X, Y split
y=dfonehot['Score Bins']
dfonehot=dfonehot.drop(['Score Bins'],axis=1)
X=dfonehot.drop('User Score', axis=1)
#Drops non numeric info
X.select_dtypes(include=np.number)
#ID is not real data, Owned is sort of recursive, and weight is a user generated score so not helpful
X.drop(columns='ID', inplace=True)
X.drop(columns='Owned', inplace=True)
X.drop(columns='Weight', inplace=True)
X.drop(columns='Year Published', inplace=True)
```

Logreg Setup

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

#we give our estimators as a list of tuples: name:function.
estimators = [('normalise', StandardScaler()),
              
              ('logreg', LogisticRegression(max_iter=500))]

pipe = Pipeline(estimators)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
pipe.fit(X_train, y_train)
print(pipe.score(X_train, y_train))
print(pipe.score(X_test, y_test))
```

```python
for importance, name in sorted(zip(pipe.named_steps['logreg'].coef_[0],X_train.columns), reverse=True):
    print(importance)
    print(name)
```
